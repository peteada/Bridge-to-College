## Bridge Course Comparison:
Compare Bridge Students vs Non-Bridge Students


```python
import pandas as pd
import numpy as np
import seaborn as sns
import time
import datetime
import csv
%matplotlib inline

source_path = 'D:\\PYTHON\\Bridge to College\\Data_Files\\' # Location of folder

source_discipline = 'BtC_Cohort 1_Discipline.csv' # Discipline file name
source_assess = 'BtC_Cohort 1_Assess.csv'
source_grades = 'BtC_Cohort 1_Grades.csv'
source_enrollment = 'BtC_Cohort 1_Enrollment.csv'
source_college = 'R3110_BERC_BTC_HEdetail.tab.D180311.T100410.ascii.T100410'

output_report = 'Output_Report_Bridge_Compare.csv' # Name of file that contains all to be used
```

```python
##### Assess file
print('Loading Assess File')
adf = pd.read_csv(source_path + source_assess)
adf['TestGrade'].replace(to_replace=['11'], value=11, inplace=True) 

##### Grades File
print('Loading Grades File')
gdf = pd.read_csv(source_path + source_grades)
gdf['StateCourseCode'].replace(to_replace=['02099'], value=2099, inplace=True)
gdf['StateCourseCode'].replace(to_replace=['01069'], value=1069, inplace=True)

##### Enrollment File
#edf = pd.read_csv(source_path + source_enrollment)

##### College FIle
cdf = pd.read_csv(source_path + source_college, sep='\t', lineterminator='\r', encoding="ISO-8859-1")   

print('All files loaded' + str(datetime.datetime.now().strftime('%Y %m %d,  %H:%M:%S')))
```

```python
##### Create the English Data Frame #####
english_df = adf[(adf['TestGrade'] == 11) &  (adf['SchoolYear'] == 2015) & 
           (adf['Subject'] == 'Smarter Balanced ELA') & 
           (adf['PerformanceLevel'] != '')][['ResearchID', 'TestType', 'PerformanceLevel', 'Subject']]
english_df.drop_duplicates(subset=None, keep='first', inplace=True)

temp_df = gdf[(gdf.SchoolYear.isin([2016])) & (gdf.StateCourseCode.isin([1069])) & 
              (gdf.GradeLevelWhenCourseTaken.isin([12]))][['ResearchID', 'StateCourseCode']]
english_df = pd.merge(english_df, temp_df, how = 'outer')
english_df.drop_duplicates(subset=None, keep='first', inplace=True)

##### College English
temp_df = cdf[(cdf['TermTTL'] == 'Fall') & 
              (cdf['PreCollegeEnglishCourseFlag'] == 'Y')][['ResearchID', 'PreCollegeEnglishCourseFlag']]
temp_df = temp_df.rename(columns = {'PreCollegeEnglishCourseFlag':'FallPreCollegeEnglish'})
english_df = pd.merge(english_df, temp_df, how = 'left', on=['ResearchID'])

temp_df = cdf[(cdf['TermTTL'] == 'Spring') & 
              (cdf['PreCollegeEnglishCourseFlag'] == 'Y')][['ResearchID', 'PreCollegeEnglishCourseFlag']]
temp_df = temp_df.rename(columns = {'PreCollegeEnglishCourseFlag':'SpringPreCollegeEnglish'})
english_df = pd.merge(english_df, temp_df, how = 'left', on=['ResearchID'])


temp_df = cdf[(cdf['TermTTL'] == 'Fall') & 
              (cdf['CollegeLevelEnglishCourseEnrollm'] == 'Y')][['ResearchID', 'CollegeLevelEnglishCourseEnrollm']]
temp_df = temp_df.rename(columns = {'CollegeLevelEnglishCourseEnrollm':'FallCollegeEnglish'})
english_df = pd.merge(english_df, temp_df, how = 'left', on=['ResearchID'])

temp_df = cdf[(cdf['TermTTL'] == 'Spring') & 
              (cdf['CollegeLevelEnglishCourseEnrollm'] == 'Y')][['ResearchID', 'CollegeLevelEnglishCourseEnrollm']]
temp_df = temp_df.rename(columns = {'CollegeLevelEnglishCourseEnrollm':'SpringCollegeEnglish'})
english_df = pd.merge(english_df, temp_df, how = 'left', on=['ResearchID'])

##### Create the Math Data Frame

math_df = adf[(adf['TestGrade'] == 11) &  (adf['SchoolYear'] == 2015) & 
           (adf['Subject'] == 'Smarter Balanced Math') & 
           (adf['PerformanceLevel'] != '')][['ResearchID', 'TestType', 'PerformanceLevel', 'Subject']]
math_df.drop_duplicates(subset=None, keep='first', inplace=True)

temp_df = gdf[(gdf.SchoolYear.isin([2016])) & (gdf.StateCourseCode.isin([2099])) & 
              (gdf.GradeLevelWhenCourseTaken.isin([12]))][['ResearchID', 'StateCourseCode']]
math_df = pd.merge(math_df, temp_df, how = 'outer')
math_df.drop_duplicates(subset=None, keep='first', inplace=True)

##### College Math 
temp_df = cdf[(cdf['TermTTL'] == 'Fall') & 
              (cdf['PreCollegeMathCourseFlag'] == 'Y')][['ResearchID', 'PreCollegeMathCourseFlag']]
temp_df = temp_df.rename(columns = {'PreCollegeMathCourseFlag':'FallPreCollegeMath'})
math_df = pd.merge(math_df, temp_df, how = 'left', on=['ResearchID'])

temp_df = cdf[(cdf['TermTTL'] == 'Spring') & 
              (cdf['PreCollegeMathCourseFlag'] == 'Y')][['ResearchID', 'PreCollegeMathCourseFlag']]
temp_df = temp_df.rename(columns = {'PreCollegeMathCourseFlag':'SpringPreCollegeMath'})
math_df = pd.merge(math_df, temp_df, how = 'left', on=['ResearchID'])

temp_df = cdf[(cdf['TermTTL'] == 'Fall') & 
              (cdf['CollegeLevelMathCourseEnrollment'] == 'Y')][['ResearchID', 'CollegeLevelMathCourseEnrollment']]
temp_df = temp_df.rename(columns = {'CollegeLevelMathCourseEnrollment':'FallCollegeMath'})
math_df = pd.merge(math_df, temp_df, how = 'left', on=['ResearchID'])

temp_df = cdf[(cdf['TermTTL'] == 'Spring') & 
              (cdf['CollegeLevelMathCourseEnrollment'] == 'Y')][['ResearchID', 'CollegeLevelMathCourseEnrollment']]
temp_df = temp_df.rename(columns = {'CollegeLevelMathCourseEnrollment':'SpringCollegeMath'})
math_df = pd.merge(math_df, temp_df, how = 'left', on=['ResearchID'])
```

```python
#############################################
#####     Functions to write row data     #####
#############################################
def write_row (group, category, num_students, num_bridge, percent_bridge, percent_precollege_fall, 
               percent_college_fall, percent_precollege_spring, percent_college_spring):
    writer.writerow({'Student_Group': group, 
                     'Category': category, 
                     'Total Students': num_students, 
                     '# of Students': num_bridge,
                     '% of Students': percent_bridge, 
                     'PreCollege Level % Fall': percent_precollege_fall, 
                     'College Level % Fall': percent_college_fall, 
                     'PreCollege Level % Spring': percent_precollege_spring,
                     'College Level % Spring': percent_college_spring})
    
#######################################################
#####            Write a simple row Row           #####
#######################################################
def write_simple_line (row_name):
    write_row(row_name, '', '', '', '', '', '', '', '')

###########################################
#####            Write  Row           #####
###########################################
def bridge_row(data_input, row_name, category):
    if data_input == 'english':
        data_frame = english_df
        FallPreCollege = 'FallPreCollegeEnglish'
        FallCollege = 'FallCollegeEnglish'
        SpringPreCollege = 'SpringPreCollegeEnglish'
        SpringCollege = 'SpringCollegeEnglish'
    else:
        data_frame = math_df
        FallPreCollege = 'FallPreCollegeMath'
        FallCollege = 'FallCollegeMath'
        SpringPreCollege = 'SpringPreCollegeMath'
        SpringCollege = 'SpringCollegeMath'
    num_students = data_frame.ResearchID.nunique()
    num_bridge = data_frame[((data_frame['StateCourseCode'] == 1069) | (data_frame['StateCourseCode'] == 2099))]['ResearchID'].nunique()
    percent_bridge = str('%4.2f' %(100 * (num_bridge / num_students)))
    
    percent_fall_precollege = str('%4.2f' %(100 * ((data_frame[((data_frame.StateCourseCode == 1069) | 
                                                               (data_frame.StateCourseCode == 2099)) & 
                                                               (data_frame[FallPreCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_bridge)))
    percent_fall_college = str('%4.2f' %(100 * ((data_frame[((data_frame.StateCourseCode == 1069) | 
                                                               (data_frame.StateCourseCode == 2099)) & 
                                                               (data_frame[FallCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_bridge)))
    percent_spring_precollege = str('%4.2f' %(100 * ((data_frame[((data_frame.StateCourseCode == 1069) | 
                                                               (data_frame.StateCourseCode == 2099)) & 
                                                               (data_frame[SpringPreCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_bridge)))
    percent_spring_college = str('%4.2f' %(100 * ((data_frame[((data_frame.StateCourseCode == 1069) | 
                                                               (data_frame.StateCourseCode == 2099)) & 
                                                               (data_frame[SpringCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_bridge)))
    
    write_row(row_name, category, num_students, num_bridge, percent_bridge, percent_fall_precollege, percent_fall_college, percent_spring_precollege, percent_spring_college)

###########################################
#####   Write  Bridge Category Row    #####
###########################################
def bridge_category_row(data_input, row_name, category, category_value):
    if data_input == 'english':
        data_frame = english_df
        FallPreCollege = 'FallPreCollegeEnglish'
        FallCollege = 'FallCollegeEnglish'
        SpringPreCollege = 'SpringPreCollegeEnglish'
        SpringCollege = 'SpringCollegeEnglish'
    else:
        data_frame = math_df
        FallPreCollege = 'FallPreCollegeMath'
        FallCollege = 'FallCollegeMath'
        SpringPreCollege = 'SpringPreCollegeMath'
        SpringCollege = 'SpringCollegeMath'

    num_students = data_frame[(data_frame[category] == category_value)]['ResearchID'].nunique()

    num_bridge = data_frame[((data_frame['StateCourseCode'] == 1069) | (data_frame['StateCourseCode'] == 2099)) &
                            (data_frame[category] == category_value)]['ResearchID'].nunique()
    percent_bridge = str('%4.2f' %(100 * (num_bridge / num_students)))
    
    percent_fall_precollege = str('%4.2f' %(100 * ((data_frame[((data_frame.StateCourseCode == 1069) | 
                                                               (data_frame.StateCourseCode == 2099)) & 
                                                               (data_frame[category] == category_value) &
                                                               (data_frame[FallPreCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_bridge)))
    percent_fall_college = str('%4.2f' %(100 * ((data_frame[((data_frame.StateCourseCode == 1069) | 
                                                            (data_frame.StateCourseCode == 2099)) & 
                                                            (data_frame[category] == category_value) &
                                                            (data_frame[FallCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_bridge)))
    percent_spring_precollege = str('%4.2f' %(100 * ((data_frame[((data_frame.StateCourseCode == 1069) | 
                                                                 (data_frame.StateCourseCode == 2099)) & 
                                                                 (data_frame[category] == category_value) &
                                                                 (data_frame[SpringPreCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_bridge)))
    percent_spring_college = str('%4.2f' %(100 * ((data_frame[((data_frame.StateCourseCode == 1069) | 
                                                              (data_frame.StateCourseCode == 2099)) & 
                                                              (data_frame[category] == category_value) &
                                                              (data_frame[SpringCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_bridge)))
    
    write_row(row_name, category_value, num_students, num_bridge, percent_bridge, percent_fall_precollege, percent_fall_college, percent_spring_precollege, percent_spring_college)

###########################################
#####   Write Non Bridge Row          #####
###########################################
def non_bridge_row(data_input, row_name, category):
    if data_input == 'english':
        data_frame = english_df
        FallPreCollege = 'FallPreCollegeEnglish'
        FallCollege = 'FallCollegeEnglish'
        SpringPreCollege = 'SpringPreCollegeEnglish'
        SpringCollege = 'SpringCollegeEnglish'
    else:
        data_frame = math_df
        FallPreCollege = 'FallPreCollegeMath'
        FallCollege = 'FallCollegeMath'
        SpringPreCollege = 'SpringPreCollegeMath'
        SpringCollege = 'SpringCollegeMath'

    num_students = data_frame.ResearchID.nunique()
    num_nonbridge = data_frame[(data_frame['StateCourseCode'] != 1069) & (data_frame['StateCourseCode'] != 2099)]['ResearchID'].nunique()
    percent_nonbridge = str('%4.2f' %(100 * (num_nonbridge / num_students)))
    
    percent_fall_precollege = str('%4.2f' %(100 * ((data_frame[(data_frame.StateCourseCode != 1069) & 
                                                               (data_frame.StateCourseCode != 2099) & 
                                                               (data_frame[FallPreCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_nonbridge)))
    percent_fall_college = str('%4.2f' %(100 * ((data_frame[(data_frame.StateCourseCode != 1069) & 
                                                               (data_frame.StateCourseCode != 2099) & 
                                                               (data_frame[FallCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_nonbridge)))
    percent_spring_precollege = str('%4.2f' %(100 * ((data_frame[(data_frame.StateCourseCode != 1069) & 
                                                               (data_frame.StateCourseCode != 2099) & 
                                                               (data_frame[SpringPreCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_nonbridge)))
    percent_spring_college = str('%4.2f' %(100 * ((data_frame[(data_frame.StateCourseCode != 1069) & 
                                                               (data_frame.StateCourseCode != 2099) & 
                                                               (data_frame[SpringCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_nonbridge)))
    
    write_row(row_name, category, num_students, num_nonbridge, percent_nonbridge, percent_fall_precollege, percent_fall_college, percent_spring_precollege, percent_spring_college)

###########################################
#####   Write Non Bridge Category Row #####
###########################################
def non_bridge_category_row(data_input, row_name, category, category_value):
    if data_input == 'english':
        data_frame = english_df
        FallPreCollege = 'FallPreCollegeEnglish'
        FallCollege = 'FallCollegeEnglish'
        SpringPreCollege = 'SpringPreCollegeEnglish'
        SpringCollege = 'SpringCollegeEnglish'
    else:
        data_frame = math_df
        FallPreCollege = 'FallPreCollegeMath'
        FallCollege = 'FallCollegeMath'
        SpringPreCollege = 'SpringPreCollegeMath'
        SpringCollege = 'SpringCollegeMath'

    num_students = data_frame[(data_frame[category] == category_value)]['ResearchID'].nunique()
    num_nonbridge = data_frame[(data_frame['StateCourseCode'] != 1069) & 
                               (data_frame['StateCourseCode'] != 2099) & 
                               (data_frame[category] == category_value)]['ResearchID'].nunique()
    percent_nonbridge = str('%4.2f' %(100 * (num_nonbridge / num_students)))
    
    percent_fall_precollege = str('%4.2f' %(100 * ((data_frame[(data_frame.StateCourseCode != 1069) & 
                                                               (data_frame.StateCourseCode != 2099) & 
                                                               (data_frame[category] == category_value) &
                                                               (data_frame[FallPreCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_nonbridge)))
    percent_fall_college = str('%4.2f' %(100 * ((data_frame[(data_frame.StateCourseCode != 1069) & 
                                                            (data_frame.StateCourseCode != 2099) & 
                                                            (data_frame[category] == category_value) &
                                                            (data_frame[FallCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_nonbridge)))
    percent_spring_precollege = str('%4.2f' %(100 * ((data_frame[(data_frame.StateCourseCode != 1069) & 
                                                                 (data_frame.StateCourseCode != 2099) & 
                                                                 (data_frame[category] == category_value) &
                                                                 (data_frame[SpringPreCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_nonbridge)))
    percent_spring_college = str('%4.2f' %(100 * ((data_frame[(data_frame.StateCourseCode != 1069) & 
                                                              (data_frame.StateCourseCode != 2099) & 
                                                              (data_frame[category] == category_value) &
                                                              (data_frame[SpringCollege] == 'Y')]
                                                    ['ResearchID'].nunique()) / num_nonbridge)))
    
    write_row(row_name, category_value, num_students, num_nonbridge, percent_nonbridge, percent_fall_precollege, percent_fall_college, percent_spring_precollege, percent_spring_college)
```

```python
########################################################
#####     Set up and Create the Comparison CSV     #####
########################################################
with open(source_path + output_report, "w", newline='') as csvfile:
    fieldnames = ['Student_Group', 'Category', 'Total Students',  '# of Students', '% of Students', 
                     'PreCollege Level % Fall', 'College Level % Fall', 'PreCollege Level % Spring', 'College Level % Spring']
                  
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
 
 ############################################################################################    
    write_simple_line('English')
    bridge_row('english', 'English Students', 'All Bridge')
    non_bridge_row('english', 'English Students', 'All non Bridge')
    bridge_category_row('english', 'English SBA Level', 'PerformanceLevel', 'L1')
    non_bridge_category_row('english', 'English Non Bridge SBA Level', 'PerformanceLevel', 'L1')
    bridge_category_row('english', 'English SBA Level', 'PerformanceLevel', 'L2')
    non_bridge_category_row('english', 'English Non Bridge SBA Level', 'PerformanceLevel', 'L2')
    bridge_category_row('english', 'English SBA Level', 'PerformanceLevel', 'L3')
    non_bridge_category_row('english', 'English Non Bridge SBA Level', 'PerformanceLevel', 'L3')
    bridge_category_row('english', 'English SBA Level', 'PerformanceLevel', 'L4')
    non_bridge_category_row('english', 'English Non Bridge SBA Level', 'PerformanceLevel', 'L4')
    
    write_simple_line('')
    write_simple_line('')
    write_simple_line('Math')
    bridge_row('math', 'Math Students', 'All Bridge')
    non_bridge_row('math', 'Math Students', 'All non Bridge')
    bridge_category_row('math', 'Math SBA Level', 'PerformanceLevel', 'L1')
    non_bridge_category_row('math', 'Math Non Bridge SBA Level', 'PerformanceLevel', 'L1')
    bridge_category_row('math', 'Math SBA Level', 'PerformanceLevel', 'L2')
    non_bridge_category_row('math', 'Math Non Bridge SBA Level', 'PerformanceLevel', 'L2')
    bridge_category_row('math', 'Math SBA Level', 'PerformanceLevel', 'L3')
    non_bridge_category_row('math', 'Math Non Bridge SBA Level', 'PerformanceLevel', 'L3')
    bridge_category_row('math', 'Math SBA Level', 'PerformanceLevel', 'L4')
    non_bridge_category_row('math', 'Math Non Bridge SBA Level', 'PerformanceLevel', 'L4')

csvfile.close()
print('Execution Completed: ' + str(datetime.datetime.now().strftime('%Y %m %d,  %H:%M:%S')))
```
